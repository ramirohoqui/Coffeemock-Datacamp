```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
coffe = pd.read_csv('coffee-exam.csv')
#check for missing values
coffe.isna().sum()
```

```python
#Columns Analysis 

#Region 
coffe['Region'].isna().sum()
#No missing Values, no edit made to column
print(coffe['Region'].unique())
```

```python
# Place Name
coffe['Place name'].isna().sum()
# No missing Values, no edit made to column
coffe['Place name'].value_counts()
```

```python
#place type
coffe['Place type'].isna().sum()
print
coffe['Place type'].value_counts()
#No missing Values, no edit made to column
```

```python
#Rating
coffe['Rating'].isna().sum()
#Replace missing values with 0
coffe['Rating'] = coffe['Rating'].fillna(0)

print(coffe['Rating'].value_counts())
#check missing values were replaced

coffe['Rating'].isna().sum()
```

```python
#Reviews
coffe['Reviews'].isna().sum()

#Replace missing values with overall median
coffe['Reviews'] = coffe['Reviews'].fillna(coffe['Reviews'].median())

print(coffe['Reviews'].value_counts())
#check missing values were replaced

coffe['Reviews'].isna().sum()
```

```python
#Price
coffe['Price'].isna().any()

#check prices format

coffe['Price'].describe()
```

```python
#Delivery Option
coffe['Delivery option'].isna().any()

print(coffe['Delivery option'].unique())
```

```python
#Dine in option

#check missing Values
print(coffe['Dine in option'].isna().sum())

#replace missing values with False

coffe['Dine in option'] = coffe['Dine in option'].fillna(False)

#check missing values
print(coffe['Dine in option'].isna().sum())
```

```python
#Takeout option

#check missing values
print(coffe['Takeout option'].isna().sum())

#replace missing values with False

coffe['Takeout option'] = coffe['Takeout option'].fillna(False)

#check missing values
                                
print(coffe['Takeout option'].isna().sum())
```

```python
#TASK 2
#Show most common types of coffe stores 
coffe.columns

sns.countplot(x='Place type',data=coffe)
plt.show()

#there are four different place types across the data. Most frequent place type is coffe shop. the categories are unbalanced with most observations either as cafe or coffee shop "place type"
```

```python
# TASK 3
# Describe distribution of all the number of reviews

import seaborn as sns

# Remove outlying values
coffe_cleaned = coffe[coffe['Reviews'] < coffe['Reviews'].quantile(0.95)]

sns.histplot(data=coffe_cleaned, x='Reviews')
plt.show()
```

```python
# TASK 4
# Relationship between type of store and number of reviews

sns.boxplot(data=coffe, x='Place type', y='Reviews')
plt.show()
```

```python
# Removing outliers
sns.boxplot(data=coffe_cleaned, x='Place type', y='Reviews')
plt.show()
```
